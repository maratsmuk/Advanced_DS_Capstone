### Model definition.
In this notebook, the models are defined and compiled. The presented functions are called only at the initial stage of the simulation.

This notebook contains only one simple function, which returns a compiled model. One of three different models can be chosen: 
1. Scikit-learn's simple linear regression.
2. Spark linear regression.
3. Dense Neural Networks (DNN's).

The first two models are simple and don't have their particular parameters at this step. The third model requires an additional parameter `inp_shape` being the input shape of the first layer. This parameter is setted automatically as `degrees+1`, where `degrees` is the number of degrees in polynomial features. 

Additional layers, the number of units for each layer, the activation functions, etc., can be modified here. The presented choice was found as the best one for the studied problem.
