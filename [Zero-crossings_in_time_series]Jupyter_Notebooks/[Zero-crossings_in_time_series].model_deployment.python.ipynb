Let us create a simple function, which encapsulates the whole simulation process for a particular test function number `fnum` ($fnum \in \{1,2,...,10\}$ for the standard test problems and $fnum \in \{-1,-2\}$ for the Bouncing ball problem) using $h0$ for the initial generation of the time observations $t_k = t_0 + h\cdot k$, where $t_0$ is generated/received from the `generate_problem` (see the `[Zero-crossings_in_time_series].data_exploration.python.ipynb` and `[Zero-crossings_in_time_series].etl.python.ipynb` notebooks for details). `N_train` is used to determine the size of the training set, while `model_context` identifies the model used during the simulation (=0 for Scikit-Learn, =1 for Spark, =2 for DNNs). This procedure returns the array of all predicted values of $g(t)$ (as `result_predicted`), all generated time observations $t$ (as `time_steps_predicted`), the found number of prediction steps (as `N_prediction`) and the used number of polynomial degrees (as `degrees`). 

```
def simulate_one_test_function(fnum,h0 = 0.0001,N_train = 1000,model_context = 0):
  #Generate the problem:
  random.seed(fnum)
  result_predicted = []
  time_steps_predicted = []
  f,MR,t0,T,sigma = generate_problem(fnum)
  T = MR+0.1
  #Generate or get the initial value in the required format:
  data1 = provide_new_time_steps(t0-0.1,1,0.1)
  data_T,data_X = provide_new_data(data1,f,sigma,model_context)
  #Generate or get remaining N_train-1 observations
  data1 = provide_new_time_steps(t0,N_train-1,h0)
  data1,data2 = provide_new_data(data1,f,sigma,model_context)
  #Find the last generated time observation and create the arrays of all time observations and their function values:
  if model_context==1:
    data_T = data_T.union(data1)   
    data_T.createOrReplaceTempView("data_T")
    t_last = spark.sql("select t from data_T order by t desc limit 1").rdd.map(lambda row: row.t.toArray()).collect()[0]
  else:
    data_T = np.append(data_T,data1)
    data_X = np.append(data_X,data2)
    t_last = data_T[-1]
  #Estimate the Lipschitz constant and return the average stepsize (useful, when a real-life problem with dynamical stepsizes is solved)
  L,h = Lipschitz_estimate(data_T,data_X,model_context,True)
  #The number of prediction observations is calculated:
  N_prediction = min(N_train,max(10,int(0.1/(h*L))))
  #N_prediction = N_train
  print("Number of prediction steps: ",N_prediction)
  #The number of degrees for polynomial features is calculated:
  degrees = find_degrees(data_T,data_X,N_train-min(int(N_train/2),N_prediction),min(int(N_train/2),N_prediction),model_context)
  #degrees = 1
  #inp_shape is required for DeepLearning algorithms only:
  if model_context==2:
    inp_shape = degrees
  else:
    inp_shape=0
  #The main simulation process: while there are time observations to predict:
  while t_last<T+h/100:
    #Generate or get time observations to predict (without their actual function values: they are predicted)
    data1 = provide_new_time_steps(t_last,N_prediction,h0)
    #Take the last N_train known observations for the training set into data_train_predict and, eventually, into data_train_X, respectively:
    if model_context==1:
      data_T.createOrReplaceTempView("data_T")
      data_train_predict = spark.sql("select * from (select * from data_T order by t desc limit "+str(N_train)+") order by t").union(provide_new_data(data1,f=lambda x: np.nan,sigma=0,model_context=1)[0])
      data_train_X = None
    else:
      data_train_predict = np.append(data_T[-N_train:],data1)
      data_train_X = data_X[-N_train:]
    #Prepare the training and prediction features:
    train_t,predict_t = prepare_features(data_train_predict,N_train,N_prediction,degrees,model_context)
    #In order to improve the performance for test problems only, deep learning models can be pre-compiled earlier. For real-life problems this is not possible. So, let's compile and fit our model:
    if model_context==2:
      try:
        #model = tf.keras.models.load_model('drive/MyDrive/Results_ADS_IBM/saved_model'+str(fnum)+'/my_model'+'_'+str(t_last)+'.h5') 
        blah_blah_just_for_fun
      except:
        model = compile_model(model_context,inp_shape)
        model = fit_model(model,train_t,data_train_X,model_context)
        #model.save('drive/MyDrive/Results_ADS_IBM/saved_model'+str(fnum)+'/my_model'+'_'+str(t_last)+'.h5') 
    else:
      model = compile_model(model_context,inp_shape)
      model = fit_model(model,train_t,data_train_X,model_context)
    #Let's predict the values:
    y_predicted = predict_model(model,predict_t,model_context)
    #Here we could check the zero-crossings:
    #Uncomment the following lines of code for this purpose:
    #RE,FZA,ZCA = evaluate_zero_crossings(data1,np.array(y_predicted))
    #if len(ZCA)>0:
    #  print("Warning! A zero-crossing has been determined! The first zero-crossing was determined approximately at t = ",FZA)
    #Append them into the final arrays of all predicted values.
    result_predicted.append(y_predicted)
    time_steps_predicted.append(data1.tolist())
    #Get or generate new data:
    data1,data2 = provide_new_data(data1,f,sigma,model_context)
    if model_context==1:
      data_T = data_T.union(data1)
      data_T.createOrReplaceTempView("data_T")
      t_last = spark.sql("select t from data_T order by t desc limit 1").rdd.map(lambda row: row.t.toArray()).collect()[0]
    else:
      data_T = np.append(data_T,data1)
      data_X = np.append(data_X,data2)
      t_last = data_T[-1]
    #Eventually, check if the model works correctly:
    #Uncomment the following lines of code for this purpose:
    if model_context==1:
      data1.createOrReplaceTempView("data1")
      t_real = spark.sql("select t from data1").rdd.map(lambda row: row.t.toArray()[0]).collect()
      y_real = spark.sql("select x from data1").rdd.map(lambda row: row.x).collect()
    else:
      t_real = data1
      y_real = data2
    #RE,FZA_1,ZCA_1 = evaluate_zero_crossings(t_real,y_real)
    #if len(ZCA_1)>0:
      #RE,FZA,ZCA = evaluate_zero_crossings(t_real,np.array(y_predicted),FZA_1)
      #break
  return result_predicted,time_steps_predicted,N_prediction,degrees
```
